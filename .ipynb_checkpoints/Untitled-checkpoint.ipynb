# learning openCV from OpenCV-python Tutorials

```python
import cv2
```

## load an image 

```python
# cv2.IMREAD_COLOR=1 ,cv2.IMREAD_GRAYSCALE=0, cv2.IMREAD_UNCHANGED=-1 
img = cv2.imread('1.jpeg', 1)
img
```

## display an image

```python
# u can resize window
cv2.namedWindow('DAD', cv2.WINDOW_NORMAL)
# fisrt arg name of window 
#second arg input
cv2.imshow('DAD', img)
# wait infinti for a key stroke
cv2.waitKey(0)
# 
cv2.destroyAllWindows()
```

## write an image

```python
cv2.imwrite('dad.png', img)
```

## using  matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
# plt.imshow(img)
plt.imshow(img, cmap = 'gray', interpolation='bicubic')
# to hide tick values on X and Y axis
plt.xticks([]), plt.yticks([])  
plt.show()
```

## change Colorspaces

```python
img1=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img1)
```

```python
red_upp = np.array([225,0,0])
red_low = np.array([225,92,92])
res = cv2.inRange(img1, red_low,red_upp)
plt.imshow(res)
```

```python
res = cv2.bitwise_and(img,img, res)
cv2.imshow('frame',res)
```
